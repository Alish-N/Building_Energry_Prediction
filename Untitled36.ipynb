```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/building_energy_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cooling_System'].fillna(df['Cooling_System'].mode()[0], inplace=True)
```

```python
df['Cooling_System'].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
categorical_features = ['Insulation_Type', 'Heating_System', 'Cooling_System']
```

```python
#Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Apply one-hot encoding to the categorical columns
one_hot_encoded = encoder.fit_transform(df[categorical_features])

#Create a DataFrame with the one-hot encoded columns
#We use get_feature_names_out() to get the column names for the encoded data
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_features))

# Concatenate the one-hot encoded dataframe with the original dataframe
df_encoded = pd.concat([df, one_hot_df], axis=1)

# Drop the original categorical columns
df_encoded = df_encoded.drop(categorical_features, axis=1)

# Display the resulting dataframe
print(f"Encoded Employee data : \n{df_encoded}")
```

```python
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
df['Average_Window_Size'] = df['Total_Area'] / df['Number_of_Windows']
```

```python
df['Total_Insulation_Area'] = df['Total_Area'] * 0.5
```

```python
df['Heating_Cooling_System_Efficiency'] = np.random.rand(len(df))
```
